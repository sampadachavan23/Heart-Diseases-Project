# Heart Diseases Prediction

```python
# Description About the Dataset :-

#heart disease
#Age: age of the patient [years]
#Sex: sex of the patient [M: Male, F: Female]
#ChestPainType: chest pain type [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]
#RestingBP: resting blood pressure [mm Hg]
#Cholesterol: serum cholesterol [mm/dl]
#FastingBS: fasting blood sugar [1: if FastingBS > 120 mg/dl, 0: otherwise]
#RestingECG: resting electrocardiogram results [Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]
#MaxHR: maximum heart rate achieved [Numeric value between 60 and 202]
#ExerciseAngina: exercise-induced angina [Y: Yes, N: No]
#Oldpeak: oldpeak = ST [Numeric value measured in depression]
#ST_Slope: the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]
#HeartDisease: output class [1: heart disease, 0: Normal]
```

# Splitting The Data

```python
x=data.drop("HeartDisease",axis=1)
y=data["HeartDisease"]
```

```python
print(x)
```

```python
print(y)
```

# Spliting the Data into training data & Test Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=12)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Model Training

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model_L = LogisticRegression()
model_L.fit(x_train,y_train)
pred_L=model_L.predict(x_test)
confusion_matrix(y_test,pred_L)
print("Accuracy score on logistic Regression is :",accuracy_score(y_test,pred_L))
```

# Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model_D=DecisionTreeClassifier()
model_D.fit(x_train,y_train)
pred_D=model_D.predict(x_test)
confusion_matrix(y_test,pred_D)
print("Accuracy score on Decision Tree model is :",accuracy_score(y_test,pred_D))
```

```python
from sklearn import tree
plt.figure(figsize=(25,20))
tree.plot_tree(model_D,filled=True,max_depth=3,fontsize=20)
plt.show()
```

# KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K=10
model_K=KNeighborsClassifier(n_neighbors=10)
model_K.fit(x_train,y_train)
pred_K=model_K.predict(x_test)
confusion_matrix(y_test,pred_K)
print("Accuracy score on KNN model is :",accuracy_score(y_test,pred_K))
```

```python
# K=12
model_K=KNeighborsClassifier(n_neighbors=12)
model_K.fit(x_train,y_train)
pred_K=model_K.predict(x_test)
confusion_matrix(y_test,pred_K)
print("Accuracy score on KNN model is :",accuracy_score(y_test,pred_K))
```

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
model_N=GaussianNB()
model_N.fit(x_train,y_train)
pred_N=model_N.predict(x_test)
confusion_matrix(y_test,pred_N)
print("Accuracy score on Naive Bayes model is :",accuracy_score(y_test,pred_N))
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model_SC=SVC()
model_SC.fit(x_train,y_train)
pred_SC=model_SC.predict(x_test)
confusion_matrix(y_test,pred_SC)
print("Accuracy score on Support Vector Machine model is :",accuracy_score(y_test,pred_SC))
```

# Random Forest Models

```python
from sklearn.ensemble import RandomForestClassifier
model_R=RandomForestClassifier()
model_R.fit(x_train,y_train)
pred_R=model_R.predict(x_test)
print("Accuracy score on Random Forest model is :",accuracy_score(y_test,pred_R))
```

```python
# Logistic Regression=0.852
# Decision Tree= 0.79
# KNN = 0.72
# Naive Bayes = 0.83
# SVM= 0.73
# RFM=0.867
# According to this Random Forest Model shows High Accuracy
```
